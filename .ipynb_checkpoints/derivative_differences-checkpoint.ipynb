```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f=pd.DataFrame()
```

$dataA=x^3+2x^2-3x-1$  
$derDataA=dataA'(x)=3x^2+4x-3$  
$der2DataA=dataA''(x)=6x+4$

```python
f['dataA']=[((x/1000)**3 + 2*(x/1000)**2 - 3*(x/1000) - 1) for x in range(-1000,2500,1)]
f['derDataA']=[(3*(x/1000)**2 + 4*(x/1000) - 3) for x in range(-1000,2500,1)]
f['der2DataA']=[(6*(x/1000) + 4) for x in range(-1000,2500,1)]
```

```python
plt.rcParams['figure.figsize'] = [12, 9]
ax=f.plot.line(y=['dataA',
               'derDataA',
               'der2DataA',
              ])
ax.axhline(y=0)
```

$dataA=-x^3+2x^2-3x-1$  
$derDataA=dataA'(x)=-3x^2+4x-3$  
$der2DataA=dataA''(x)=-6x+4$

```python
f['dataA']=[(-(x/1000)**3 + 2*(x/1000)**2 - 3*(x/1000) - 1) for x in range(-1000,2500,1)]
f['derDataA']=[(-3*(x/1000)**2 + 4*(x/1000) - 3) for x in range(-1000,2500,1)]
f['der2DataA']=[(-6*(x/1000) + 4) for x in range(-1000,2500,1)]
```

```python
f.plot.line(y=['dataA',
               'derDataA',
               'der2DataA'
              ])
```

$dataA=-x^3-2x^2-3x-1$  
$derDataA=dataA'(x)=-3x^2-4x-3$  
$der2DataA=dataA''(x)=-6x-4$

```python
f['dataA']=[(-(x/1000)**3 - 2*(x/1000)**2 - 3*(x/1000) - 1) for x in range(-1000,2500,1)]
f['derDataA']=[(-3*(x/1000)**2 - 4*(x/1000) - 3) for x in range(-1000,2500,1)]
f['der2DataA']=[(-6*(x/1000) - 4) for x in range(-1000,2500,1)]
```

```python
f.plot.line(y=['dataA',
               'derDataA',
               'der2DataA'
               ])
```

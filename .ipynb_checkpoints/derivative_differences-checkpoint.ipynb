```python
import pandas as pd
```

```python
f=pd.DataFrame()
```

$dataA=2x^2-3x-1$  
$derDataA=dataA'(x)=4x-3$  
$der2DataA=dataA''(x)=4$

```python
f['dataA']=[(2*(x/1000)**2 - 3*(x/1000) - 1) for x in range(-1000,2500,1)]
f['derDataA']=[(4*(x/1000) - 3) for x in range(-1000,2500,1)]
f['der2DataA']=[(4) for x in range(-1000,2500,1)]
```

```python
f.plot.line(y=['dataA',
               'derDataA',
               'der2DataA'
              ])
```
